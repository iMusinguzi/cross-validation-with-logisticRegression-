```python
#cross validation, on iris data
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
```

```python
iris = load_iris()
lr = LogisticRegression()
```

```python
score = cross_val_score(lr,iris.data,iris.target)
print('Cross-validation scores: {}'.format(score))
```

```python
#we can summarize the cv accuracy by computing the mean
print('Average cross-validation score: {}'.format(score.mean()))
```

```python
#we can have more control over cross vslidation by adjusting the number of folds
from sklearn.model_selection import KFold
kfold = KFold(n_splits= 5)
print('Cross validation scores: \n{}'.format(cross_val_score(lr,iris.data,
                                                            iris.target,cv=kfold)))
```

```python
# we can also include shuffle parameter so that we remove 
# the ordering of the samples
kfold = KFold(n_splits = 3,shuffle=True,random_state=0)
print('Cross-validation score:\n{}'.format(cross_val_score(lr,iris.data,
                                                           iris.target,cv=kfold)))
```

```python
#think of leave-one-out as k-fold cv where each fold is a single sample
from sklearn.model_selection import LeaveOneOut
loo = LeaveOneOut()
score = cross_val_score(lr,iris.data,iris.target,cv=loo)
print("Number of cv iterations: ",len(score))
print('Mean accuracy:{:.2f}'.format(score.mean()))
```

```python
#Shuffle-split cross validation 
#it allows control over the number of iterations
from sklearn.model_selection import ShuffleSplit
ss = ShuffleSplit(test_size=.5, train_size=.5, n_splits=10)
score = cross_val_score(lr,iris.data,iris.target,cv=ss)
print('Cross-validation scores:\n{}'.format(score))
```

There are more splitting strategies for cross-validation in scikit-learn, which
allow for an even greater variety of usecases. The standard KFold,StratifiedKFold, and GroupFKold are by far the most commonly used ones
